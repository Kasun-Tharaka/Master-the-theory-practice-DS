## Methods of Lists

```python
fruits.append('pears') # add into end of the list
```

```python
print(fruits)
```

```python
fruits.insert(1, "coconut") # insert into the index mentioned, others shifted and same
print(fruits)
```

```python
fruits.remove('orange')
print(fruits)
```

```python
popedfruits = fruits.pop() # remove the last element of the list and return it
print(fruits)
```

```python
index = fruits.index('melon') # return the index of the element mentioned
print(index)
```

```python
fruits.insert(4, 'apple')
print(fruits.count('apple'))
```
