While Loop

```python
# it countinus to execute until the condition is false

count = 0
 
while count < 5:
    print(count)
    count += 1 
```

```python
# chech even numbers

count = 10

while count%2 == 0:
    print(count)
    count += 1
```

```python
for i in range(count):
    if i%2 == 0:
        print(i)
```

```python
# break and continue

for i in range(20):
    if i == 7:
        break
    print(i)
```

```python
for i in range(10):
    if i == 7:
        continue # skip the current item and continiue with the next item
    print(i)
```
