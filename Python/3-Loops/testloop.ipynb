While Loop

```python
# it countinus to execute until the condition is false

count = 0
 
while count < 5:
    print(count)
    count += 1 
```

```python
# chech even numbers

count = 10

while count%2 == 0:
    print(count)
    count += 1
```

```python
for i in range(count):
    if i%2 == 0:
        print(i)
```

```python
# break and continue

for i in range(20):
    if i == 7:
        break
    print(i)
```

```python
for i in range(10):
    if i == 7:
        continue # skip the current item and continiue with the next item
    print(i)
```

```python
# print only odd numbers
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
# nested loops
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}")
```

```python
n = 10
sum = 0
count = 1

while count <=n:
    sum = sum + count
    count = count + 1

print(sum)
```

```python
sum = 0

for i in range(12):
    sum = sum + i
print(sum)
```

```python
# prime numbers 1 and 100

for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```
