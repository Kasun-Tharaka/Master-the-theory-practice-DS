```python
# Read a Whole File

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

```python
# this code overwrite the file completely
with open('example.txt', 'w') as file:
    file.write('this is new to file line')
```

```python
with open('example.txt', 'a') as file:
    file.write('\nthis is appended line to the file')
```

```python
with open('example.txt', 'r') as file:
    for i in file:
        print(i.strip())
```
